### **Gráficos con Matplotlib**

Matplotlib es una biblioteca de visualización de datos en Python.  Es una de las herramientas más populares y versátiles para crear gráficos estáticos, animados e interactivos en dos dimensiones.  Proporciona una amplia gama de opciones de personalización, lo que la hace adecuada para una variedad de propósitos, desde simples gráficos hasta visualizaciones complejas.


Algunos puntos clave sobre Matplotlib:

* **Funcionalidad:** Permite crear una gran variedad de gráficos, incluyendo líneas, barras, histogramas, dispersión, contornos, imágenes, etc. También ofrece herramientas para personalizar la apariencia de los gráficos, como colores, etiquetas, leyendas, títulos y anotaciones.

* **Modularidad:**  Su arquitectura modular permite a los usuarios acceder a diferentes partes de la biblioteca según sus necesidades.  Por ejemplo, `pyplot` es un módulo de interfaz de alto nivel que proporciona una interfaz similar a MATLAB, mientras que `axes` ofrece un nivel más granular de control.

* **Integración:** Se integra fácilmente con otras bibliotecas de Python, como NumPy y Pandas, lo que facilita la creación de visualizaciones a partir de datos almacenados en arrays o DataFrames.

* **Personalización:** Ofrece una amplia gama de opciones de personalización para el estilo y apariencia de los gráficos, permitiendo a los usuarios ajustarlos a sus necesidades específicas.

* **Tipos de gráficos:**

    * **Gráficos de líneas:** Ideales para mostrar tendencias y cambios en datos a lo largo del tiempo.
    * **Gráficos de barras:** Útiles para comparar cantidades discretas.
    * **Histogramas:** Representan la distribución de frecuencia de datos continuos.
    * **Gráficos de dispersión:** Muestran la relación entre dos variables.
    * **Gráficos de sectores:** Representan porcentajes de un conjunto de datos.

* **Plataformas:** Matplotlib puede generar gráficos en varios formatos, como PNG, JPG, PDF, SVG, etc., lo que los hace adecuados para diferentes medios e imprimibles.


```python
""" Gráficos con Matplotlib """
# Importamos la biblioteca Matplotlib.pyplot:
import matplotlib.pyplot as plt
# Importamos numpy, util para el gráfico de histograma:
import numpy as np
```

```python
""" Gráfico de líneas """
# Datos de ejemplo para nuestro gráfico de líneas:
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# LLamamos la función para plot y le pasamos
# los ejes en el orden de los ejes x - y,
plt.plot(x, y)

# Se imprime la imagen:
plt.show()
```

```python
""" Gráfico de líneas con estilo y etiquetas """
# Datos de ejemplo para el gŕafico de lineas:
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# Cambiar color, estilo y añadir marcadores
plt.plot(x, y, color='red', linestyle='--', marker='o')

# Titulo del gráfico
plt.title('Mi primer gráfico de líneas')

# Etiquetas de los ejes:
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadir una cuadrícula
plt.grid(True)
plt.show()
```

```python
""" Gráfico de barras """
# Datos de ejemplo para nuestro gráfico de barras:
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [20, 35, 30, 25, 15]

# Cambiar el color de las barras:
plt.bar(categorias, valores, color='green')
plt.title('Ventas por Categoría')

plt.xlabel('Categorías')
plt.ylabel('Ventas')
plt.show()
```

```python
""" Gráfico de barras con estilo """
# Datos de ejemplo para nuestro gráfico de barras:
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [20, 35, 30, 25, 15]

# Cambiar el color de las barras:
plt.bar(categorias, valores, color='skyblue')
plt.title('Ventas por Categoría')

plt.xlabel('Categorías')
plt.ylabel('Ventas')

 # Añadir una cuadrícula solo en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
""" Gráfico histograma """
# Datos para gráfico:
datos = np.random.normal(0, 1, 1000)

 # Cambiar color y bordes
plt.hist(datos, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de Datos Aleatorios')

plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
""" Gráfico de torta """
# Datos para gráfico
categorias = ['A', 'B', 'C', 'D']
valores = [30, 20, 35, 15]

# Definir colores de los sectores
colores = ['skyblue', 'lightgreen', 'lightcoral', 'gold']

# Añadir porcentajes, colores y ángulo inicial
plt.pie(valores, labels=categorias, colors=colores, autopct='%1.1f%%', startangle=90)

# Añadir titulo
plt.title('Distribución de Categorías')

# Mostrar
plt.show()
```

```python
# prompt: Gemini puedes explicarme como personalizar los gráficos de las celdas anteriores, con ejemplos, y explicando para que se debe usar cada uno de estos gráficos

import matplotlib.pyplot as plt
import numpy as np

# Ejemplo de gráfico de líneas con personalización avanzada
x = np.linspace(0, 10, 100)  # Datos en el eje x
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura

plt.plot(x, y1, label='Seno', color='blue', linewidth=2, linestyle='--', marker='o', markersize=5, markerfacecolor='red')
plt.plot(x, y2, label='Coseno', color='green', linewidth=2, linestyle='-', marker='s', markersize=5, markerfacecolor='orange')

plt.title('Gráfico de Funciones Trigonométricas', fontsize=16, fontweight='bold') # Título con estilo
plt.xlabel('Eje X', fontsize=12)
plt.ylabel('Eje Y', fontsize=12)

plt.legend(loc='upper right', fontsize=10) # Leyenda ubicada en la esquina superior derecha
plt.grid(True, linestyle=':', alpha=0.6)

plt.xlim(0, 10) # Límite de los ejes x e y
plt.ylim(-1.5, 1.5)

plt.xticks(np.arange(0, 11, 1)) # Personalización de las marcas en el eje x

plt.text(5, 0.8, 'Punto de Intersección', fontsize=10, color='purple', bbox=dict(facecolor='lightblue', alpha=0.5)) # Añadir texto en la gráfica

# Guardar gráfico en archivo (opcional)
# plt.savefig('grafico_personalizado.png', dpi=300)

plt.show()


# Ejemplo de histograma con personalización
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))

plt.hist(data, bins=50, color='skyblue', alpha=0.7, edgecolor='black', linewidth=1.2) #  bins = número de contenedores del histograma
plt.title('Histograma de Datos Aleatorios', fontsize=14)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.show()

# Ejemplo de gráfico de barras con personalización
categorias = ['A', 'B', 'C', 'D']
valores = [25, 37, 15, 20]

plt.figure(figsize=(8, 10))

plt.bar(categorias, valores, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'], width=0.6) # Ancho de las barras
plt.title('Gráfico de Barras Personalizado', fontsize=14)
plt.xlabel('Categorías', fontsize=12)
plt.ylabel('Valores', fontsize=12)

# Agregar etiquetas de valor encima de cada barra
for i, v in enumerate(valores):
    plt.text(i, v + 1, str(v), ha='center', va='bottom') # Añadir etiquetas de texto a cada barra

plt.show()
```

### Gráficos con Seaborn.

Seaborn es una biblioteca de visualización de datos en Python basada en matplotlib.  Proporciona una interfaz de alto nivel para crear gráficos estadísticos atractivos e informativos.  Se usa para visualizar relaciones entre variables, distribuciones de datos, comparaciones entre grupos y tendencias a lo largo del tiempo, entre otras cosas.

En esencia, Seaborn facilita la creación de gráficos más complejos y estéticamente agradables que con Matplotlib básico, especialmente cuando se trabaja con conjuntos de datos estadísticos.  Automatiza muchos aspectos de la personalización visual, lo que reduce la cantidad de código necesaria para generar gráficos de alta calidad.

Aquí te dejo algunas de sus ventajas:

* **Interfaz de alto nivel:** Simplifica la creación de gráficos complejos.  Con menos líneas de código puedes generar gráficos sofisticados.
* **Gráficos estadísticos:** Está diseñado específicamente para visualizar datos estadísticos. Incluye funciones para gráficos de dispersión, histogramas, gráficos de violín, diagramas de caja, mapas de calor y mucho más.
* **Integración con Pandas:** Trabaja a la perfección con DataFrames de Pandas, lo que facilita el análisis y la visualización de datos tabulares.
* **Personalización:** Permite un alto grado de personalización de la apariencia de los gráficos, aunque de manera más sencilla que Matplotlib puro.
* **Estilos y paletas de colores:** Ofrece estilos y paletas de colores predefinidos que mejoran la estética de los gráficos de forma predeterminada.


En resumen, si bien Matplotlib es la base, Seaborn lo potencia para análisis de datos estadísticos, ofreciendo una forma más eficiente y visualmente atractiva de explorar tus datos.


```python
""" Gráficos con Seaborn """
# Importamos la biblioteca Seaborn:
import seaborn as sns
```

```python
""" Gráfico de líneas con Seaborn """
# Datos de ejemplo para nuestro gráfico de líneas:
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

sns.lineplot(x=x, y=y)
plt.title('Gráfico de Líneas con Seaborn')
plt.show()
```

```python
""" Gráfico de barras con Seaborn """
# Datos de ejemplo para nuestro gráfico de barras:
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [20, 35, 30, 25, 15]

sns.barplot(x=categorias, y=valores)
plt.title('Gráfico de Barras con Seaborn')
plt.show()
```

```python
""" Gráfico Histograma con Seaborn """
# Datos
datos = np.random.normal(0, 1, 1000)

sns.histplot(datos, bins=30)
plt.title('Histograma con Seaborn')
plt.show()
```

```python
""" Gráfico de dispersión con Seaborn """
# Datos de ejemplo para el gráfico de dispersión
x = np.random.rand(50)
y = np.random.rand(50)

sns.scatterplot(x=x, y=y)
plt.title('Gráfico de Dispersión con Seaborn')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

# **Casos de uso de los gráficos presentados:**

### **Gráfico de líneas:**
1. Mostrar tendencias a lo largo del tiempo (ej: precios de acciones, temperatura diaria, crecimiento de usuarios).
2. Comparar el comportamiento de varias variables en función del tiempo (ej: rendimiento de dos estrategias de inversión).
3. Visualizar series temporales para identificar patrones, ciclos o anomalías.


### **Gráfico de líneas con estilo y etiquetas:**
Igual que el anterior, pero con mayor énfasis en la comunicación efectiva.  Las etiquetas, colores y estilos ayudan a destacar información relevante y facilitar la interpretación.


### **Gráfico de barras:**
1. Comparación de cantidades discretas entre diferentes categorías (ej: ventas por producto, población por país, resultados de encuestas).
2. Mostrar la distribución de una variable categórica (ej: número de clientes por segmento).


### **Gráfico de barras con estilo:**
Mejora la legibilidad y atractivo visual del gráfico de barras, facilitando la comparación de datos.


### **Gráfico de histograma:**
1. Visualizar la distribución de una variable continua (ej: alturas de una población, calificaciones de un examen).
2. Identificar la forma de la distribución (normal, sesgada, etc.) y detectar valores atípicos.
3. Comprender la dispersión y centralidad de los datos.


### **Gráfico de torta:**
1. Mostrar la proporción de diferentes partes de un todo (ej: composición de un presupuesto, participación de mercado).
2. Representar porcentajes de forma visualmente atractiva.
3. Ideal para datos categóricos que representan proporciones.


### **Gráfico de líneas con personalización avanzada:**
Ideal para reportes o publicaciones donde la estética y precisión son críticas.  Permite una presentación profesional de datos con múltiples series y detalles relevantes, usando diferentes estilos y colores para distinguir entre series, títulos informativos, leyendas que indican qué representa cada línea y un texto con un punto clave para destacar.

### **Histograma con personalización:**
Mejorar la visualización de un histograma para destacar aspectos importantes de la distribución de los datos.


### **Gráfico de barras con personalización:**
Permite una presentación más informativa al añadir etiquetas a cada barra, mostrando los valores de forma directa y mejorando la legibilidad del gráfico. Ideal para reportes donde la claridad y precisión son esenciales.
